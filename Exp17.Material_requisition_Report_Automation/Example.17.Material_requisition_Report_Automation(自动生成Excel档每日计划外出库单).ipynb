# 实例17：自动生成Excel档每日计划外出库单

除了工单领料，公司还有一种领料方式，是直接开领料单领取，叫做“计划外出库”。一年的计划外出库单太多，包含所有部门的，那要挑出研发部的，非常困难。因此只得从系统中导出研发部领取的记录，然后再还原成纸本的“计划外出库单”。因此需要从总表中拆分。此例与例16类似，只不过表位的设置稍微复杂一点点，可以作为练手的材料。具体细节可参考例16，就不做重复解释了。

总表和每天的表分别长这样的：
![](images\rawdata.png)
![](images\result.png)

```python
#1.从总表中提取信息
from openpyxl import load_workbook
wb = load_workbook("data\研发材料-计划外出库.xlsx")
ws= wb.active
data = {} #用于储存提取的信息
for row in range(2, ws.max_row+1): #从第2行开始（第1行是标题）遍历工作表每一行，将数据提取出来
    material_pn = ws['A' + str(row)].value #为物料编号
    material_name = ws['B' + str(row)].value #为物料名称
    material_desp = ws['C' + str(row)].value #为物料中文名称
    qty= ws['F' + str(row)].value #为数量 
    sub= ws['G' + str(row)].value #为库位
    reason_code = "B2" 
    reason = "研发用料"
    FG_pn =  ws['J' + str(row)].value
   
    date = ws['K' + str(row)].value.date() #为日期时间，因只要日期，不要时间，所以用date()只提取日期
    info_list=[material_pn,material_name,material_desp,qty,sub,reason_code,reason,FG_pn] #将以上信息放入列表info_list
    data.setdefault(date,[]) #data字典将以日期date作为键，当天的所有信息组成的列表嵌套列表作为值
    data[date].append(info_list) #将单个物料信息的列表放入包含所有物料的大列表
```

数据获取完成后，就可以开始创建并写入“计划外出库”清单了。

```python
%%time
#2.将提取的信息按日期写入新建的Excel表
from openpyxl.styles import Border, Side, PatternFill, Font, GradientFill, Alignment
wb_day = load_workbook("data\计划外出库调整单模板.xlsx")
ws_day= wb_day.active

thin = Side(border_style="thin", color="000000")#定义边框粗细及颜色

for date in data.keys():
    ws_new = wb_day.copy_worksheet(ws_day) #复制模板中的工作表
    ws_new.title=str(date)[-5:] #以日期为新工作表命名
    i=6 #计数器，从第6行开始填写
    seq=1 #物料种类计数，初始为1
    
# ['49999RM0004900', 'DOUBLE-SIDED ADHESIVE TAPE 48MM*15Y', 
#   '双面胶 48MM*15Y', -30, 'WH', 'B2', '研发用料', 'R1PA257A00CP00001'], 
    
    for material in data[date]: #获取每天的每个物料信息，逐个写入工作表
        ws_new.cell(row=i,column=1).value=seq
        ws_new.cell(row=i,column=2).value=material[0]
        ws_new.cell(row=i,column=3).value=material[1]
        ws_new.cell(row=i,column=4).value=material[2]#第5、6列为空，直接跳到第7列
        ws_new.cell(row=i,column=7).value=str(material[3]).split(".")[0]#砍掉小数点后的部分
        ws_new.cell(row=i,column=8).value=material[4]
        ws_new.cell(row=i,column=9).value=material[5]
        ws_new.cell(row=i,column=10).value=material[6]
        ws_new.cell(row=i,column=11).value=material[7]

        
        i+=1 #每写一行，计数器就需要加1，以便从下一行接着写入
        seq+=1
    i=i+2
    ws_new.cell(row=i,column=2).value="提出部门："
    ws_new.cell(row=i,column=2).font= Font(bold=True,size=10)
    ws_new.cell(row=i+1,column=2).value="日  期："
    ws_new.cell(row=i+1,column=2).font= Font(bold=True,size=10)
    ws_new.cell(row=i,column=3).value="工程部"
    ws_new.cell(row=i,column=3).font= Font(size=10)
    ws_new.cell(row=i,column=3).border = Border(bottom=thin)
    ws_new.cell(row=i+1,column=3).value=date
    ws_new.cell(row=i+1,column=3).font= Font(size=10)
    ws_new.cell(row=i+1,column=3).alignment = Alignment(horizontal="left", vertical="center")
    ws_new.cell(row=i+1,column=3).border = Border(bottom=thin)
    ws_new.cell(row=i,column=5).value="货仓确认："
    ws_new.cell(row=i,column=5).font = Font(bold=True,size=10)
    ws_new.cell(row=i,column=6).border = Border(bottom=thin)
    ws_new.cell(row=i+1,column=5).value="物控部确认："
    ws_new.cell(row=i+1,column=5).font = Font(bold=True,size=10)
    ws_new.cell(row=i+1,column=6).border = Border(bottom=thin)
    ws_new.cell(row=i+2,column=1).value="备注：1.账务调整由货仓、物控主管签名确认生效！"
    ws_new.cell(row=i+2,column=1).font = Font(bold=True,size=10)
    ws_new.cell(row=i+3,column=2).value="2.成品计划外出入库，可凭相关邮件或文件，货仓主管签名生效！"
    ws_new.cell(row=i+3,column=2).font = Font(bold=True,size=10)
    ws_new.cell(row=i+3,column=11).value="XX-OP-991"
    ws_new.cell(row=i+3,column=11).font = Font(size=9)
    # 调整“签字行”行高
    ws_new.row_dimensions[i].height = 25
    ws_new.row_dimensions[i+1].height = 25
    
    #设置字号，对齐，缩小字体填充，加边框

    for row_number in range(6, 5+seq):
        for col_number in range(1,12):
            c = ws_new.cell(row=row_number,column=col_number)
            c.font = Font(size=8)
            c.border = Border(top=thin, left=thin, right=thin, bottom=thin)
            c.alignment = Alignment(horizontal="left", vertical="center",wrap_text = True)


wb_day.save("data\计划外出库领料单（每日）.xlsx")
```
